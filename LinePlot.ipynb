```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Cargamos el csv que guarda el fitness de la solución actual de cada iteración del algoritmo local search.
#El fichero incluye las ejecuciones que optimizan la instancia "bur26a".
df = pd.read_csv('./Ejecuciones/iterations.csv',header=None)

df.columns=["Instance","Functions","Iteration","Fitness"]

#Dado que solo queremos fijarnos en la convergencia del algoritmo, ignoramos las primeras 20 iteraciones.
df = df[df["Iteration"]>20]

#Analizamos por separado las tres configuraciones de parámetros que hemos utilizado para el algoritmo.
configuration = [" 1 -1 -1 "," 2 -1 -1 "," 3 -1 -1 "]
position = ["1","2","3"]

#Imprimimos los resultados en forma de gráfico. Las líneas indican el fitness medio de cada iteración del
#local search para las tres configuraciones de parámetros utilizadas. El área sombreada indica la desviación 
#estándar del fitness para cada caso (en base a la muestra de ejecuciones).
plt.figure(figsize=(15,12))
for i in range(len(configuration[:1])):
    df_function = df[df["Functions"]==configuration[i]]
    sns.lineplot(data=df, x=df_function["Iteration"], y=df_function["Fitness"], ci='sd',legend="brief",label=position[i])

plt.title("BUR26A",fontsize=40)
plt.show()
```
